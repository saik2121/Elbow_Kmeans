```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#To create a dataset of 500 points with 5 centers having a feature of 2
X,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
plt.figure(0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
Kmeans=KMeans(n_clusters=1,init='k-means++')
Kmeans.fit(X)
```

```python
Kmeans.inertia_
```

```python
inertia_list=[]
for cluster in range(1,20):
    Kmeans=KMeans(n_clusters=cluster,init='k-means++')
    Kmeans.fit(X)
    inertia_list.append(Kmeans.inertia_)
    
```

```python
data=pd.DataFrame({'cluster':range(1,20),'inertia':inertia_list})
plt.plot(data['cluster'],data['inertia'],marker='*',color='red')
plt.figure(figsize=(15,6))
plt.show()
```
